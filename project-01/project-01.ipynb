Your predicted values in this case are your "predictions" of what the true, observed values are.

Your function will:

#### Define your function below:

````python
import numpy as np
def root_mean_squared_error(true_values, predictions):
    #function that calculates the root mean squared error,
    errors = [pred - val for pred, val in zip(predictions, true_values)]
    sq_errors = [ i ** 2 for i in errors]
    avg_sq_error = np.mean(sq_errors)
    rmse = np.sqrt(avg_sq_error)
    print rmse
    return rmse
````
